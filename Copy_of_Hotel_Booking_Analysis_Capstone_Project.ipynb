**NUMBER OF BOOKING VS COUNTRY**

 we can see that  customers come from Portugal, Great Britain, France and Spain
 has highest number of booking 

```python
grp_by_country = hd.groupby('country')
s1 = pd.DataFrame(grp_by_country.size()).rename(columns = {0:'no. of bookings'}).sort_values('no. of bookings', ascending = False)
s1 = s1[:10]
plt.figure(figsize = (10,7))
sns.barplot(x = s1.index, y = s1['no. of bookings'])
plt.show()
```

```python
country_wise_guests = hd[hd['is_canceled'] == 0]['country'].value_counts().reset_index()
country_wise_guests.columns = ['country', 'No of guests']
country_wise_guests
```

**CITY HOTEL VS RESORT HOTEL IN TERMS OF BOOKING**

The below graph shows us that customers are more interested in city hotel as compare to resort hotel 
it clearly shows that the city hotel has 66% of the booking and resort hotel has almost 33% of the booking 

```python
grouped_by_hotel = hd.groupby('hotel')
d1 = pd.DataFrame((grouped_by_hotel.size()/hd.shape[0])*100).reset_index().rename(columns = {0:'Booking %'})      #Calculating percentage
plt.figure(figsize = (8,5))
sns.barplot(x = d1['hotel'], y = d1['Booking %'] )
plt.show()
```

**CITY HOTEL VS RESORT HOTEL IN TERMS OF REVENUE**

The below graphe shows that the city hotel is gernerating more revenue as compare to the resort hotel 

```python
d3 = grouped_by_hotel['adr'].agg(np.mean).reset_index().rename(columns = {'adr':'avg_adr'})   # calculating average adr
plt.figure(figsize = (8,5))
sns.barplot(x = d3['hotel'], y = d3['avg_adr'] )
plt.show()
```

**CHANNEL MOSTLY USED FOR EARLY BOOKING HOTELS**

The below grapha shows us that TA/TO is mostly used for planning Hotel visits ahead of time. But for sudden visits other mediums are most preferred.




```python
group_by_dc = hd.groupby('distribution_channel')
d3 = pd.DataFrame(round(group_by_dc['lead_time'].median(),2)).reset_index().rename(columns = {'lead_time': 'median_lead_time'})
plt.figure(figsize = (10,5))
sns.barplot(x = d3['distribution_channel'], y = d3['median_lead_time'])
plt.show()
```

**Booking cancellation Analysis**

The distribution channels i.e TA/TO has large cancelation percentage as compared to other distribution channels 



```python
group_by_dc = hd.groupby('distribution_channel')
d1 = pd.DataFrame(round((group_by_dc.size()/hd.shape[0])*100,2)).reset_index().rename(columns = {0: 'Cancel_%'})
plt.figure(figsize = (8,8))
data = d1['Cancel_%']
labels = d1['distribution_channel']
plt.pie(x=data, autopct="%.2f%%", explode=[0.05]*5, labels=labels, pctdistance=0.5)
plt.title("Cancel_% by distribution channels", fontsize=14);
```

 **AVERAGE WAITING TIME VS HOTELS**

 Waiting time period for the city hotel is high as compared to resort hotels. this means that city hotel is the most preferred hotel by the customers and is the busiest hotels 

```python
d5 = pd.DataFrame(grouped_by_hotel['days_in_waiting_list'].agg(np.mean).reset_index().rename(columns = {'days_in_waiting_list':'avg_waiting_period'}))
plt.figure(figsize = (8,5))
sns.barplot(x = d5['hotel'], y = d5['avg_waiting_period'] )
plt.show()
```

```python
hd_copy= hd.copy()
hd_copy['no_of_bookings']= 1
hd_copy.loc[(hd_copy.arrival_date_month== 'July')| (hd_copy.arrival_date_month== 'August'),'no_of_bookings']/=3
hd_copy.loc[~((hd_copy.arrival_date_month== 'July')| (hd_copy.arrival_date_month== 'August')),'no_of_bookings']/=2
hd_copy.sample(100)
```

**HOTEL WISE BOOKING YEARLY**

The comparison between city hotel and resort hotel shows us that the booking in the  city hotel is increasing across the years then the resort hotel and it did not increase proportionatly over the years

we can also say that the no of booking are more in the year 2016 as compared to 2015 & 2017

```python
sns.countplot (x= 'arrival_date_year', data= hd, hue= 'hotel').set_title ('yearly bookings')
```

```python
waiting_bookings = hd[hd['days_in_waiting_list'] !=0]  # Selecting bookings with non zero waiting time


fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.kdeplot(ax=axes[0],x = 'days_in_waiting_list', hue = 'is_canceled' , data = waiting_bookings)
sns.kdeplot(ax = axes[1], x = hd['lead_time'], hue = hd['is_canceled'])
plt.show()
```

We see that most of the bookings that are cancelled have waiting period of less 150 days but also most of bookings that are not cancelled also have waiting period less than 150 days. Hence this shows that waiting period has no effect on cancellation of bookings.  Also, lead time has no effect on cancellation of bookings
